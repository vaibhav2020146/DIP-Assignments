```python
def operation(k,sigma,row,col):
  for x in range(-k, k + 1):
    for y in range(-k, k + 1):
        kernel[x + k, y + k] = (1 / (2 * np.pi * sigma**2)) * np.exp(-(x**2 + y**2) / (2 * sigma**2))

  for i in range(k, row - k):
    for j in range(k, col - k):
      output_img[i, j] = int(np.sum(img[i - k:i + k + 1, j - k:j + k + 1] * kernel))
```

```python
def display(output_img):
  cv2_imshow(output_img)
  cv2.waitKey()
  cv2.destroyAllWindows()
```
