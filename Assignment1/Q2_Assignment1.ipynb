```python
from google.colab.patches import cv2_imshow
from skimage.io import imread,imshow
import cv2
import numpy as np
```

```python
def operation(row,col,kernel,out,img):
  for i in range(1,row-1):#traversing the image
    for j in range(1,col-1):
      var=0
      for var1 in range(-1,2):
        for var2 in range(-1,2):
          var+=img[i+var1,j+var2]*kernel[var1+1,var2+1]#multiplying the kernel with the image pixels so that we can get the desired output
          out[i,j]=var#storing output in a variable
```

```python
def display(output_image):
  cv2_imshow(output_image)
  cv2.waitKey()
  cv2.destroyAllWindows()
```

# First - order Techniques (Sobel)


```python
img=imread('tank.jpg',0)
kernel = np.array([[0,0,0 ],
 [0,0,-1 ],
 [0,1 ,0]])
row,col=img.shape
out=np.zeros([row,col])
operation(row,col,kernel,out)
row,col=img.shape
kernel_different = np.array([[-1,-1,1 ],
 [0,0,0 ],
 [1,1 ,1]])
output=np.zeros([row,col])
for i in range(1,row-1):
 for j in range(1,col-1):
  var=0
  for var1 in range(-1,2):
   for var2 in range(-1,2):
    var+=img[i+var1,j+var2]*kernel[var1+1,var2+1]
    output[i,j]=var
output_image=out+output#adding the two outputs because in sobel technique we have to add the two outputs because we have to find the gradient of the image in both the directions
display(output_image)
```
